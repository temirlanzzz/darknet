```python
with open('results.txt','r') as f:
    l = f.readline()
    i = 1
    sum = 0.0
    while l:
        sum += float(l.split(' ')[0].split('=')[1])
        if i % 21 == 0:
            print('inference# ',i//21, sum)
        i += 1
        l = f.readline()
```
